```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm


# for this project i have used both numpy arrays and lists using unique features from both of them
# 
#  (numpy is typically faster than lists and used more in scientific computing in python)

print("Choose an option:")
choice = input("1. Show me the bell!\n2. I want to choose parameters by myself.")

N_toss = 100000
sample_size =20
N_samples = 10000

if int(choice) == 2:
    N_toss = int(input("Number of tosses: "))
    sample_size = int(input("Sample size: "))
    N_samples = int(input("Number of samples: "))

L1 = np.random.randint(0, 2, N_toss)

L2 = []
for i in range(N_samples):
    L2.append(np.random.choice(L1, size=sample_size).sum())

plt.figure(figsize=(10,8))

plt.hist(L2, bins=15, density=True)

mu, std = norm.fit(L2) 
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)
plt.plot(x, p, 'k', linewidth=2)
plt.show()
```
